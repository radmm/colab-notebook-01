```python
import cv2
import numpy as np
import pandas as pd
import supervision as sv
from ultralytics import YOLO
from collections import defaultdict
import google.generativeai as genai
from google.colab import userdata
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import StandardScaler
import plotly.express as px
import warnings
warnings.filterwarnings("ignore")
```

```python
genai.configure(api_key=userdata.get("GEMINI_API_KEY"))
gemini = genai.GenerativeModel("gemini-1.5-flash")

MODEL_PATH = "yolov8x-pose.pt"          # downloads automatically on first run
VIDEO_PATH = "/content/clip.mp4"         # upload your clip here
OUTPUT_PATH = "/content/output.mp4"
ASYMMETRY_THRESHOLD = 10.0               # % — above this is flagged
LANDING_ANGLE_NORMAL = (20, 25)          # degrees
SPEED_DECAY_THRESHOLD = 10.0            # % drop
```

```python
from ultralytics import YOLO

model = YOLO(MODEL_PATH)

tracker = sv.ByteTrack()
box_annotator = sv.BoxAnnotator(thickness=2)
edge_annotator = sv.EdgeAnnotator(color=sv.Color.GREEN, thickness=2)
label_annotator = sv.LabelAnnotator(text_scale=0.5)
```

```python
def angle_between(a, b, c):
    """
    Compute angle at point b formed by a-b-c.
    a, b, c are (x, y) keypoint coordinates.
    Returns angle in degrees.
    """
    a, b, c = np.array(a), np.array(b), np.array(c)
    ba = a - b
    bc = c - b
    cosine = np.dot(ba, bc) / (np.linalg.norm(ba) * np.linalg.norm(bc) + 1e-6)
    return np.degrees(np.arccos(np.clip(cosine, -1.0, 1.0)))


def compute_knee_asymmetry(keypoints):
    """
    YOLOv8-pose keypoint indices:
      11 = left hip, 13 = left knee, 15 = left ankle
      12 = right hip, 14 = right knee, 16 = right ankle
    Returns asymmetry percentage between left and right knee bend angles.
    """
    kp = keypoints  # shape: (17, 2)
    try:
        left_angle  = angle_between(kp[11], kp[13], kp[15])
        right_angle = angle_between(kp[12], kp[14], kp[16])
        asymmetry   = abs(left_angle - right_angle) / ((left_angle + right_angle) / 2 + 1e-6) * 100
        return round(asymmetry, 2), round(left_angle, 2), round(right_angle, 2)
    except Exception:
        return 0.0, 0.0, 0.0


def compute_landing_angle(keypoints):
    """
    Approximate landing impact angle using ankle-knee-hip alignment.
    Deviation from vertical = stress on joint.
    """
    kp = keypoints
    try:
        left  = angle_between(kp[15], kp[13], kp[11])
        right = angle_between(kp[16], kp[14], kp[12])
        return round((left + right) / 2, 2)
    except Exception:
        return 0.0


def compute_speed_decay(position_history):
    """
    Given list of (x, y) center positions across frames,
    compare avg speed in first half vs second half.
    Returns % speed decay.
    """
    if len(position_history) < 10:
        return 0.0
    positions = np.array(position_history)
    displacements = np.linalg.norm(np.diff(positions, axis=0), axis=1)
    mid = len(displacements) // 2
    first_half  = displacements[:mid].mean() + 1e-6
    second_half = displacements[mid:].mean()
    decay = (first_half - second_half) / first_half * 100
    return round(max(decay, 0.0), 2)
```

```python
player_data = defaultdict(lambda: {
    "positions": [],
    "asymmetry_scores": [],
    "landing_angles": [],
    "frames": []
})

cap     = cv2.VideoCapture(VIDEO_PATH)
fps     = cap.get(cv2.CAP_PROP_FPS)
width   = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height  = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
fourcc  = cv2.VideoWriter_fourcc(*"mp4v")
out     = cv2.VideoWriter(OUTPUT_PATH, fourcc, fps, (width, height))

frame_idx = 0

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    results    = model(frame, verbose=False)[0]
    detections = sv.Detections.from_ultralytics(results)
    detections = tracker.update_with_detections(detections)

    keypoints_all = results.keypoints.xy.cpu().numpy()  # (n_players, 17, 2)

    labels = []

    for i, tracker_id in enumerate(detections.tracker_id):
        if tracker_id is None:
            continue

        pid = int(tracker_id)
        kp  = keypoints_all[i]

        # Position (bounding box center)
        x1, y1, x2, y2 = detections.xyxy[i]
        cx, cy = (x1 + x2) / 2, (y1 + y2) / 2
        player_data[pid]["positions"].append((cx, cy))

        # Features
        asym, l_knee, r_knee = compute_knee_asymmetry(kp)
        landing               = compute_landing_angle(kp)

        player_data[pid]["asymmetry_scores"].append(asym)
        player_data[pid]["landing_angles"].append(landing)
        player_data[pid]["frames"].append(frame_idx)

        # Risk flag for label
        risk = "⚠" if asym > ASYMMETRY_THRESHOLD else "✓"
        labels.append(f"P{pid} {risk}")

    # Annotate frame
    frame = box_annotator.annotate(frame, detections)
    frame = label_annotator.annotate(frame, detections, labels=labels)
    try:
        frame = edge_annotator.annotate(frame, sv.KeyPoints.from_ultralytics(results))
    except Exception:
        pass

    out.write(frame)
    frame_idx += 1

cap.release()
out.release()
print(f"Processed {frame_idx} frames. Output saved to {OUTPUT_PATH}")

```

```python
records = []

for pid, data in player_data.items():
    if len(data["positions"]) < 5:
        continue

    avg_asymmetry  = np.mean(data["asymmetry_scores"])
    max_asymmetry  = np.max(data["asymmetry_scores"])
    avg_landing    = np.mean(data["landing_angles"])
    speed_decay    = compute_speed_decay(data["positions"])
    frame_count    = len(data["frames"])

    records.append({
        "player_id":       pid,
        "avg_asymmetry":   round(avg_asymmetry, 2),
        "max_asymmetry":   round(max_asymmetry, 2),
        "avg_landing_angle": round(avg_landing, 2),
        "speed_decay":     speed_decay,
        "frames_tracked":  frame_count,
    })

features_df = pd.DataFrame(records)
print(features_df)
```

```python
def risk_score(row):
    score = 0
    if row["avg_asymmetry"] > ASYMMETRY_THRESHOLD:
        score += 35
    if row["max_asymmetry"] > ASYMMETRY_THRESHOLD * 1.5:
        score += 15
    if row["avg_landing_angle"] > LANDING_ANGLE_NORMAL[1]:
        score += 30
    if row["speed_decay"] > SPEED_DECAY_THRESHOLD:
        score += 20
    return min(score, 100)

features_df["risk_score"] = features_df.apply(risk_score, axis=1)
features_df["risk_level"] = features_df["risk_score"].apply(
    lambda s: "HIGH" if s >= 60 else ("MEDIUM" if s >= 30 else "LOW")
)

print(features_df[["player_id", "risk_score", "risk_level"]].sort_values("risk_score", ascending=False))

```

```python
def gemini_observation(row):
    prompt = f"""
You are a sports medicine analyst. A player was tracked during an NBA game clip.

Movement data:
- Average knee asymmetry: {row['avg_asymmetry']}% (flag above 10%)
- Peak knee asymmetry: {row['max_asymmetry']}%
- Average landing impact angle: {row['avg_landing_angle']} degrees (normal range: 20-25)
- Speed decay across clip: {row['speed_decay']}% (flag above 10%)
- Risk score: {row['risk_score']}/100

Write exactly 2 sentences. First sentence: what the data suggests biomechanically.
Second sentence: what a coach should do about it.
Be direct, clinical, no fluff.
"""
    try:
        response = gemini.generate_content(prompt)
        return response.text.strip()
    except Exception as e:
        return f"Gemini error: {e}"


features_df["gemini_obs"] = features_df.apply(gemini_observation, axis=1)

for _, row in features_df.iterrows():
    print(f"\n--- Player {row['player_id']} | Risk: {row['risk_level']} ({row['risk_score']}/100) ---")
    print(row["gemini_obs"])
```
